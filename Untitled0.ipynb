````python
import pandas as pd
import json
import re

file_path = '/content/drive/MyDrive/github/모두의말뭉치/resource/data/nikluge-ea-2023-train.jsonl'

data = []
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        data.append(json.loads(line))
df = pd.DataFrame(data)

# 컬럼 조정
df['기쁨'] = df['output'].apply(lambda x: x['joy'] )
df['기대'] = df['output'].apply(lambda x: x['anticipation'] )
df['신뢰'] = df['output'].apply(lambda x: x['trust'] )
df['놀람'] = df['output'].apply(lambda x: x['surprise'] )
df['혐오'] = df['output'].apply(lambda x: x['disgust'] )
df['두려움'] = df['output'].apply(lambda x: x['fear'] )
df['분노'] = df['output'].apply(lambda x: x['anger'] )
df['슬픔'] = df['output'].apply(lambda x: x['sadness'] )

# input 컬럼을 form 과 target으로 분리
df['form'] = df['input'].apply(lambda x : x['form'])
df['target'] = df['input'].apply(lambda x : x['target'])

# 'form'컬럼에서 '&'가 1개만 존재하는 행에 대해서만 '그리고'로 대체
df['form'] = df['form'].apply(lambda x : x.replace('&', ' 그리고 ') if x.count('&') == 1 else x)
df['form'] = df['form'].apply(lambda x : x[:x.index('&')] + x[x.rindex('&')+1:] if x.count('&') == 2 else x)

# 'input' 컬럼을 한글, 띄어쓰기, 숫자를 제외한 텍스트를 'etc' 컬럼에 저장
df['etc_form'] = df['form'].apply(lambda x: re.sub(r'[^\w\s\d가-힣]', '', x))

# 삭제된 텍스트를 추출하여 'etc_deleted' 컬럼에 저장
# df['etc_deleted'] = df['input'].apply(lambda x: re.sub(r'[\w\s\d가-힣]', '', x))

del df['id']
del df['output']
del df['input']

# 결과 확인
df
````
